```python
import numpy as np
from matplotlib import pyplot as plt
from neural_net import neural_net # My neural net
from sklearn import datasets
import matplotlib.pyplot as plt
import time
```

### Get digits dataset and preprocess

```python
digits = datasets.load_digits()
X = np.matrix(digits.data)
z = list(digits.target)
A = np.eye(10)
y = np.matrix(list(map(lambda i: A[i], z)))
```

### Some illustrations...

```python
plt.figure(figsize=(15,15))
for i in range(16):
    plt.subplot(4, 4, i+1)
    plt.axis('off')
    plt.title('Image {0}, target={1}\nvec={2}'.format(i, z[i], y[i]))
    plt.imshow(X[i].reshape(8,8), cmap=plt.cm.gray)
plt.show()
```

### Initialize and train the neural net

```python
N = neural_net(nodes_vector = [64, 15, 10])
```

```python
costs = []
num_minutes = 5 # Train for this many minutes.
timeout = time.time() + 60*num_minutes
while time.time() < timeout:
    N.train(X, y)
    costs.append(N.cost(X, y))
```

```python
plt.plot(costs)
plt.xlabel('training iterations')
plt.ylabel('cost')
plt.show()
```

### Check accuracy on training data

```python
def predict(data):
    outs = N.h(data)
    vec = np.matrix([[i] for i in range(10)])
    maxes = np.max(outs, 1)
    preds = np.floor(outs/maxes)
    z = preds * vec
    return np.array(z.flatten())[0]
```

```python
accuracy = np.sum(predict(X) == np.array(z)) / len(z)
accuracy
```
